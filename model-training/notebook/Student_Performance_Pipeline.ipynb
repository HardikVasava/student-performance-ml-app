```python
target_mat = 'G3_mat'
```

```python
features_mat = [
    'sex', 'age', 'studytime_mat', 'failures_mat', 'absences_mat',
    'goout_mat', 'health_mat', 'G1_mat', 'G2_mat'
]
```

```python
target_por = 'G3_por'
```

```python
features_por = [
    'sex', 'age', 'studytime_por', 'failures_por', 'absences_por',
    'goout_por', 'health_por', 'G1_por', 'G2_por'
]
```

```python
df_mat = df[features_mat + [target_mat]].copy()
df_por = df[features_por + [target_por]].copy()
```

```python
df_mat = pd.get_dummies(df_mat, drop_first=True)
df_por = pd.get_dummies(df_por, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_mat = df_mat.drop(columns=target_mat)
y_mat = df_mat[target_mat]
X_train_mat, X_test_mat, y_train_mat, y_test_mat = train_test_split(X_mat, y_mat, test_size=0.3, random_state=42)
```

```python
X_por = df_por.drop(columns=target_por)
y_por = df_por[target_por]
X_train_por, X_test_por, y_train_por, y_test_por = train_test_split(X_por, y_por, test_size=0.3, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV, ShuffleSplit
from sklearn.metrics import mean_squared_error, make_scorer
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
rmse_scorer = make_scorer(rmse, greater_is_better=False)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
import lightgbm as lgb
```

```python
def find_best_model_using_randomized_search(X, y):
    algos = {
        'ridge': {
            'model': Ridge(),
            'params': {'model__alpha': [0.1, 1.0, 10.0]}
        },
        'lasso': {
            'model': Lasso(max_iter=10000),
            'params': {'model__alpha': [0.001, 0.01, 0.1, 1]}
        },
        'random_forest': {
            'model': RandomForestRegressor(),
            'params': {'model__n_estimators': [100, 200], 'model__max_depth': [5, 10, None]}
        },
        'gradient_boosting': {
            'model': GradientBoostingRegressor(),
            'params': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.01, 0.1], 'model__max_depth': [3, 5]}
        },
        'xgboost': {
            'model': XGBRegressor(objective='reg:squarederror', verbosity=0),
            'params': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.01, 0.1], 'model__max_depth': [3, 5]}
        },
        'lightgbm': {
            'model': lgb.LGBMRegressor(verbosity=-1),
            'params': {'model__n_estimators': [100, 200], 'model__learning_rate': [0.01, 0.1], 'model__num_leaves': [15, 31]}
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {'model__max_depth': [5, 10, None]}
        },
        'svr': {
            'model': SVR(),
            'params': {'model__C': [0.1, 1, 10], 'model__kernel': ['rbf', 'linear']}
        },
        'knn': {
            'model': KNeighborsRegressor(),
            'params': {'model__n_neighbors': [3, 5, 7]}
        },
        'ada_boost': {
            'model': AdaBoostRegressor(),
            'params': {'model__n_estimators': [50, 100], 'model__learning_rate': [0.01, 0.1, 1]}
        }
    }

    results = []
    cv = ShuffleSplit(n_splits=3, test_size=0.3, random_state=42)

    for name, config in algos.items():
        print(f"\nTraining: {name}")
        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('model', config['model'])
        ])
        search = RandomizedSearchCV(pipeline, config['params'], n_iter=10, scoring=rmse_scorer, cv=cv, verbose=0, random_state=42, n_jobs=-1)
        search.fit(X, y)
        best_rmse = -search.best_score_
        print(f"{name} Best RMSE: {best_rmse:.4f}")
        results.append({
            'model': name,
            'best_rmse': best_rmse,
            'best_params': search.best_params_,
            'best_estimator': search.best_estimator_
        })

    return pd.DataFrame(results)
```

```python
print("Math Models Training and Evaluation:")
results_mat = find_best_model_using_randomized_search(X_train_mat, y_train_mat)
results_mat = results_mat.sort_values(by='best_rmse')
print(results_mat)
```

```python
best_idx_mat = results_mat['best_rmse'].idxmin()
best_model_mat = results_mat.loc[best_idx_mat, 'best_estimator']
best_model_name_mat = results_mat.loc[best_idx_mat, 'model']
print(f"Selected Best Math Model: {best_model_name_mat}")
```

```python
print("\nPortuguese Models Training and Evaluation:")
results_por = find_best_model_using_randomized_search(X_train_por, y_train_por)
results_por = results_por.sort_values(by='best_rmse')
print(results_por)
```

```python
best_idx_por = results_por['best_rmse'].idxmin()
best_model_por = results_por.loc[best_idx_por, 'best_estimator']
best_model_name_por = results_por.loc[best_idx_por, 'model']
print(f"Selected Best Portuguese Model: {best_model_name_por}")
```

```python
y_pred_mat = best_model_mat.predict(X_test_mat)
final_rmse_mat = rmse(y_test_mat, y_pred_mat)
print(f"Math Test RMSE: {final_rmse_mat:.4f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test_mat, y_pred_mat, alpha=0.5)
plt.plot([0, 20], [0, 20], color='red', lw=2)
plt.xlabel('Actual G3 (Math)')
plt.ylabel('Predicted G3 (Math)')
plt.title(f'Predicted vs Actual Final Grades (Math - {best_model_name_mat})')
plt.show()
```

```python
y_pred_por = best_model_por.predict(X_test_por)
final_rmse_por = rmse(y_test_por, y_pred_por)
print(f"Portuguese Test RMSE: {final_rmse_por:.4f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test_por, y_pred_por, alpha=0.5)
plt.plot([0, 20], [0, 20], color='red', lw=2)
plt.xlabel('Actual G3 (Portuguese)')
plt.ylabel('Predicted G3 (Portuguese)')
plt.title(f'Predicted vs Actual Final Grades (Portuguese - {best_model_name_por})')
plt.show()
```
